```python
import pandas as pd

players = "../data/raw/atp_players.csv"
ranking = "../data/raw/atp_rankings_"
```

```python
# primeiro, vamos juntar o ranking "atual" no ranking 20's

ranking_20s = ranking + "20s.csv"
current = ranking + "current.csv"

df = pd.read_csv(ranking_20s)
current_df = pd.read_csv(current)

df = pd.concat([df, current_df], ignore_index=True)

df.to_csv(ranking_20s, index=False)
```

```python
# concat de todos os csv's dos rankings
csv_final = "../data/clean/atp_rankings_00-20s.csv"

df = pd.DataFrame()

for decade in ["00", "10", "20"]:
    old_df = pd.read_csv(ranking + decade + "s.csv")

    df = pd.concat([df, old_df], ignore_index=True)

df.to_csv(csv_final, index=False)
```

```python
# converter as datas no ranking  para ano-mes-dia
csv = "../data/clean/atp_rankings_00-20s.csv"

df = pd.read_csv(csv)

df["ranking_date"] = pd.to_datetime(df["ranking_date"], format="%Y%m%d")

df.to_csv(csv, index=False)
```

```python
# merge dos csv's dos rankings com o atp_players.csv

csv_rankins = "../data/clean/atp_rankings_00-20s.csv"
csv_players = "../data/raw/atp_players.csv"

df_rankings = pd.read_csv(csv_rankins)
df_players = pd.read_csv(csv_players, low_memory=False)

df_rankings = df_rankings.rename(columns={"player": "player_id"})

df_merged = df_rankings.merge(df_players[["player_id", "name_first", "name_last", "dob"]], on="player_id", how="left")

df_merged.to_csv(csv_rankins, index=False)
```

```python
# converter as datas no atp players para ano-mes-dia
csv = "../data/clean/atp_rankings_00-20s.csv"

df = pd.read_csv(csv)

df["dob"] = pd.to_datetime(df["dob"], format="%Y%m%d")

df.to_csv(csv, index=False)
```
